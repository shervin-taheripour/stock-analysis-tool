```python
# 1. Data Overview
# Sources: Alpha Vantage and yfinance
# Tickers: AAPL (both), MSFT (yfinance)
# Date Range: 2023 and last 6 months

import sys
from datetime import datetime, timedelta

# --- Extend sys.path for local modules ---
sys.path.extend([
    '../ingestion/alpha_vantage',
    '../ingestion/yfinance'
])

# --- Alpha Vantage Imports ---
from alpha_v_methods import av_get_daily_history, av_get_close

# --- yfinance Imports ---
from yfinance_methods import yf_get_data, yf_get_adj_close

# --- Date Config ---
end_date = datetime.today().date()
start_date = end_date - timedelta(days=180)

# --- Helper Function ---
def print_data_overview(df, name="DataFrame"):
    print(f"\n🗂️ {name}")
    print("-" * 40)
    print(df.info())
    print(df.describe().T)
    display(df.head())

# --- Load Alpha Vantage Data ---
df_av_full = av_get_daily_history("AAPL", env_path="../ingestion/.env", output_size="compact")
df_av_close = av_get_close("AAPL", env_path="../ingestion/.env", output_size="compact")

# --- Load yfinance Data ---
df_yf_full = yf_get_data("AAPL", start="2023-01-01", end="2023-12-31")
df_yf_close = yf_get_adj_close("MSFT", start=start_date, end=end_date)

# --- Display Summaries ---
print_data_overview(df_av_full, "Alpha Vantage - Full OHLCV")
print_data_overview(df_av_close, "Alpha Vantage - Close Only")
print_data_overview(df_yf_full, "yfinance - Full OHLCV")
print_data_overview(df_yf_close, "yfinance - Adjusted Close (MSFT)")
```
